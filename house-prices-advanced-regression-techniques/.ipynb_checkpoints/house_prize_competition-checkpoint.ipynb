### 2.2 Replace high null-containing categorical columns by binary values

Let's view the Fence column, because of the high null rate

```python
train_df.Fence.value_counts()
```

```python
train_df.Fence.dtype
```

It seems like this is a categorical column with high amount of null values. To still use the information included here, we have to convert it in a binary column (House has a Fence or Not).

```python
train_df.Fence.value_counts().sum()
```

```python
train_df.Fence.apply(lambda x: 1 if type(x) == str else 0).value_counts()
```

```python
train_df.Fence = train_df.Fence.apply(lambda x: 1 if type(x) == str else 0)
```

```python
train_df.Fence.dtype
```

We successfully changed the __Fence__ column to a binary column with dtype int64. We also checked the value counts of the non-null values vs. the null-values to ensure data integrity.

There is also a second column which shows a high amount of null values: __FireplaceQu__
<br> So let's explore it.

```python
train_df.FireplaceQu.value_counts()
```

```python
train_df.FireplaceQu.dtype
```

It seems like this is also a categorical column with high amount of null values. To still use the information included here, we have to convert it in a binary column (House has a FireplaceQu or Not) accordingly to the Fence column above.

```python
train_df.FireplaceQu.value_counts().sum()
```

```python
train_df.FireplaceQu.apply(lambda x: 1 if type(x) == str else 0).value_counts()
```

```python
train_df.FireplaceQu = train_df.FireplaceQu.apply(lambda x: 1 if type(x) == str else 0)
```

```python
train_df.FireplaceQu.dtype
```

We also successfully changed the __FireplaceQu__ column to a binary column with dtype int64. We also checked the value counts of the non-null values vs. the null-values to ensure data integrity.

### 2.3 Fill null values of columns with low amount of null

Let's go through the dataframe columns from left to right. At first we view the column __LotFrontage__.

```python
train_df.LotFrontage.dtype
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(train_df.LotFrontage.value_counts(), bins=50, kde=True)
plt.show()
```

__Idea__: If we find a highly correlated column to this one we can simply drop it, because we have redundant information.

```python
corr_train_df = train_df.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(corr_train_df[(corr_train_df > 0.75) | (corr_train_df < -0.75)], vmin=-1.0, vmax=1.0, cmap='coolwarm')
```

The column LotFrontage is __not__ highly correlated with another column. But we found column which are highly correlated with each other: 
* __OverallQual__ <-> __SalePrice__
* __GarageYrBlt__ <-> __YearBuilt__
* __TotRmsAbvGrd__ <-> __GrLivArea__
* __1stFlrSF__ <-> __TotalBsmtSF__
* __Fireplaces__ <-> __FireplaceQu__
* __GarageArea__ <-> __GarageCars__

Keep the __LotFrontage__ column in mind for further analysis.

Get the intersection of the highly correlated and null containing columns of the dataframe

```python
# get highly correlated column names as list
high_corr_cols = corr_train_df[(corr_train_df < 1.0) & ((corr_train_df > 0.75) | (corr_train_df < -0.75))]\
.dropna(axis=1, how='all').columns.tolist()
high_corr_cols
```

```python
# get nan containing column names as list
nan_cols = train_df.columns[train_df.isna().any()].tolist()
nan_cols
```

```python
# get intersection
list(set(high_corr_cols) & set(nan_cols))
```

We can safely remove the __GarageYrBlt__ column due to high correlation and containing null values.

```python
train_df.drop('GarageYrBlt', axis=1, inplace=True)
```

Let's drop the __FireplaceQu__ column, because we converted it to binary column and it highly correlates to the __Fireplaces__ column (yields better quality information).

```python
train_df.drop('FireplaceQu', axis=1, inplace=True)
```

From the highly correlated columns are only the following left:
* __TotRmsAbvGrd__ <-> __GrLivArea__
* __1stFlrSF__ <-> __TotalBsmtSF__
* __GarageArea__ <-> __GarageCars__

The relation __OverallQual <-> SalePrice__ is removed from feature engineering, because the pair contains the target varible __SalePrice__.

Let's drop the columns __TotRmsAbvGrd__, __1stFlrSF__ and __GarageCars__ due to high correlated redundant information

```python
train_df.drop(['TotRmsAbvGrd', '1stFlrSF', 'GarageCars'], axis=1, inplace=True)
```
